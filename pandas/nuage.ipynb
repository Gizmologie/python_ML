## NUAGE DE POINT = ETUDIER UNE CORRELATION

Il faut 33 éléments mini pour qu'une stats soit viable.
Une corrélation est "validé" à partir de 60%

```python
import pandas as pd 
import numpy as np 
import matplotlib as mpt
import seaborn as sns

sns.set_theme()
```

```python
df = pd.read_csv('./data/house_pricing.csv');
df
#df.info()
```

```python
nuage = df[df["SalePrice"]<500000]
```

```python
sns.relplot(data=nuage[nuage["LotArea"]<20000], x="LotArea", y="SalePrice")
```

```python
dataF = df[df["LotFrontage"]<200]
```

```python
# Autre facon de faire :
hp_data = pd.read_csv("./data/house_pricing.csv");
sns.scatterplot(data=hp_data[(hp_data["LotArea"]< 20000)&(hp_data["SalePrice"]< 500000)], x="LotArea", y="SalePrice")
```

```python
sns.relplot(data=dataF[dataF['LotArea']<100000], y="LotFrontage", x="LotArea")
```
