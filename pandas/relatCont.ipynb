```python
import pandas as pd 
import numpy as np 
import matplotlib as mpt
import seaborn as sns

sns.set_theme()
```

```python
df = pd.read_csv('./data/sales_predictions.csv');
df
```

```python
sns.relplot(data=df, x="date", y="item_price")
```

```python
ds = df.sample(50)
```

```python
sns.relplot(data=ds, x="date", y="item_price").fig.set_size_inches(20,6)
```

```python
ds["date"] = pd.to_datetime(ds['date'])
sns.relplot(data=ds, x="date", y="item_price").fig.set_size_inches(20,6)
```
